```python
df2 = df.groupby('RelationshipSatisfaction')[['WorkLifeBalance']] .mean()
df2.plot(kind = 'bar')
```

Nesse gráfico de barras é possível observar que não necessariamente quanto maior a satisfação do funcionário com o trabalho, melhor é o seu equilíbrio entre a sua vida pessoal e emprego.

```python
sns.scatterplot(data=df, x='Attrition', y='YearsSinceLastPromotion')
plt.xlabel('Attrition')
plt.ylabel('YearsSinceLastPromotion')
plt.show()
```

Pode-se observar nesse gráfico de dispersão a relação entre o tempo desde a última promoção e a probabilidade de afastamento do funcionário da empresa.

```python
sns.scatterplot(data = df, x = 'DistanceFromHome', y = 'DailyRate')
plt.xlabel('DistanceFromHome')
plt.ylabel('DailyRate')
plt.show()
```

Nesse gráfico de dispersão há a relação entre a distância da casa dos funcionários até a empresa e quanto um funcionário recebe por dia de trabalho.

```python
sns.boxplot(data = df, x = 'StockOptionLevel', y = 'Attrition')
plt.xlabel('StockOptionLevel')
plt.ylabel('Attrition')
plt.show()
```

Nesse gráfico boxplot é possível entender como as opções de ação das empresas influenciam na decisão de afastamento dos funcionários.

# **Insights**

Com essa análise podemos tirar as seguintes conclusões:

1. O equilíbrio entre a vida pessoal dos funcionários com o trabalho não tem uma influência direta na sua satisfação com a empresa, mas depende de algo provavelmente interno do ambiente de trabalho.

2. Existem pessoas insatisfeitas com a situação da empresa mesmo tendo recebido uma promoção recente.

3. A distância entre a casa dos funcionários e o trabalho influencia diretamente no seu recebimento diário, visto que quem mora mais longe obtém o gasto maior com transporte.

4. O nível de opções de ações da empresa contribui para a decisão de afastamento dos funcionários.
